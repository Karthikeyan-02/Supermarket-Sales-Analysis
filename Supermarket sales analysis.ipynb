```python
import pandas as pd
import matplotlib.pyplot as plt
def Show_Citywise(df):
    col1 = "Branch"          
    col2 = "Product line"  
    for city in df[col1].unique():
        temp_df = df[df[col1] == city]
        result = temp_df.groupby(col2)['Quantity'].sum().sort_values()
        result.plot(kind='barh', color='orange', edgecolor='black')
        plt.title(f"{city} - Total Quantity by {col2}", fontsize=14)
        plt.xlabel("Total Quantity", fontsize=12)
        plt.ylabel(col2, fontsize=12)
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
df = pd.read_csv("supermarket_sales.csv")
Show_Citywise(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("supermarket_sales.csv")
df['Date'] = pd.to_datetime(df['Date'])
monthly_data = [df[df['Date'].dt.month == month]['Total'].values for month in range(1,12+1)]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.boxplot(monthly_data)
plt.xticks(ticks=range(1,13), labels=months)
plt.title("Boxplot With monthwise collection")
plt.xlabel("Month")
plt.ylabel("Total")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("supermarket_sales.csv")

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Group data month-wise for boxplot
monthly_data = [df[df['Date'].dt.month == month]['Total'].values for month in range(1, 13)]

# Plot
plt.figure(figsize=(10, 6))
plt.boxplot(monthly_data)

# Set custom month labels
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=range(1, 13), labels=month_labels)

plt.title("Boxplot of Monthly Collection (Total Sales)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
